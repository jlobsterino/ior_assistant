# Отчёт «Финансовые последствия по ИОР» (v2 — новая схема БЗ)

Формирует Excel-отчёт по финансовым последствиям ИОР за указанный период. Один ряд = одно финансовое последствие ( `fin_impact_sid` ).

**Источники (новая схема БЗ):**
* `arnsdpsbx_t_team_sva_oarb_4.d6_base_of_knowledge_ior` — основная таблица инцидента
* `arnsdpsbx_t_team_sva_oarb_4.d6_base_of_knowledge_incident_fin_impact` — детализация фин. последствий

**Изменения относительно v1:**
* Финансовые последствия теперь в отдельной таблице `incident_fin_impact` (17 полей)
* Поле `fin_impact_subkind_name` удалено
* `fin_impact_local_crncy_code` содержит корректную локальную валюту (BYN/RUB)
* `fin_impact_type_name` присутствует
* Поле `fin_impact_ctgry_risk_code` в новой БЗ отсутствует -> флаги `fin_credit_risk_flag` / `fin_market_risk_flag` не используются в этом отчёте. Кредитная/некредитная разбивка по инциденту в целом — через готовые поля main (`incdnt_drct_dmg_cred_rub_amt`, `incdnt_drct_dmg_noncred_rub_amt` и т.п.)
* Добавлены поля `fi_busn_area_id`, `fi_org_struct_id` — направление деятельности и оргструктура на уровне последствия
* Флаги риска инцидента — string Y/N (напрямую из main)

**Запуск:** Kernel -> Restart Kernel and Run All Cells

## 1. Параметры

```python
# --- Схема и таблицы
DM_NAME          = 'arnsdpsbx_t_team_sva_oarb_4.'
MAIN_TABLE       = 'd6_base_of_knowledge_ior'
FIN_IMPACT_TABLE = 'd6_base_of_knowledge_incident_fin_impact'

# --- Период формирования отчёта
incdnt_entry_dt_begin = '2025-01-20'
incdnt_entry_dt_end   = '2025-01-30'
```

## 2. Инициализация Spark

```python
import os
import sys
import datetime
import pandas as pd

from pyspark import SparkContext, SparkConf, HiveContext
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql import functions as F, types as T
from pyspark.sql.types import StringType, DateType
from pyspark.sql.functions import (
    col, split as split_, lower as Low, upper as Up, when, lit, concat_ws, date_add,
    count, substring, regexp_replace, sum as sum_, lag, posexplode, to_timestamp,
    month, round as round_, mean as mean_, stddev as stddev_, sqrt as sqrt_
)

_ts = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
# DataLab: /tmp иногда не writable. Spark shuffle/temp файлы -> $HOME.
import os as os
_SPARK_TMP = os.path.expanduser('~/.spark-local-dir')
os.makedirs(_SPARK_TMP, exist_ok=True)
os.environ['SPARK_LOCAL_DIRS'] = _SPARK_TMP

# DataLab: Spark поднимается локально в контейнере пользователя (Local[*]),
# а не submit'ится на YARN. spark.port.maxRetries=100 - на случай коллизий
# портов между несколькими пользователями JupyterHub.
conf = SparkConf().setAppName('financial_consequences_ior_report_v2_' + _ts)
conf.setAll(
    [
        ("spark.ui.enabled",            "true"),
        ("spark.master",                "local[*]"),
        ("spark.executor.cores",        "2"),
        ("spark.executor.memory",       "8g"),
        ("spark.executor.memoryOverhead", "1g"),
        ("spark.driver.memory",         "8g"),
        ("spark.driver.maxResultSize",  "8g"),
        ("spark.port.maxRetries",       "100"),
        ("spark.local.dir",             _SPARK_TMP),
    ]
)

spark = SparkSession.builder.config(conf=conf).enableHiveSupport().getOrCreate()

# Arrow для zero-copy Spark->Pandas (в 5-10х быстрее обычного toPandas)
spark.conf.set('spark.sql.execution.arrow.pyspark.enabled', 'true')
spark.conf.set('spark.sql.execution.arrow.maxRecordsPerBatch', '50000')
print(f'Spark {spark.version} запущен.')
```

## 3. Подготовка данных

```python
print('Подготовка данных...')

_UNICODE_GARBAGE = r'[\u1128-\uFFFF\x02\x08\x0b]'
_DT_BEGIN = F.to_timestamp(F.lit(incdnt_entry_dt_begin), 'yyyy-MM-dd')
_DT_END = F.date_add(F.to_timestamp(F.lit(incdnt_entry_dt_end), 'yyyy-MM-dd'), 1)

# --- Основная таблица ИОР за период
main_table = (
    spark.table(DM_NAME + MAIN_TABLE)
    .filter(
        (F.col('incdnt_entry_dt') >= _DT_BEGIN)
        & (F.col('incdnt_entry_dt') < _DT_END)
    )
)

# --- Детали фин. последствий
fin_impact = (
    spark.table(DM_NAME + FIN_IMPACT_TABLE)
    .select(
        F.col('incdnt_id').alias('fi_incdnt_id'),
        'fin_impact_id', 'fin_impact_sid',
        'fin_impact_kind_name',
        'fin_impact_rub_amt', 'fin_impact_ccy_amt', 'fin_impact_local_ccy_amt',
        'fin_impact_crncy_code', 'fin_impact_local_crncy_code',
        'fin_impact_monitoring_flag',
        'fin_impact_detection_dt', 'fin_impact_creation_dttm', 'fin_impact_reg_dt',
        'fin_impact_account_num', 'fin_impact_docum_num',
    )
)

enriched = main_table.join(
    fin_impact,
    F.col('incdnt_id') == F.col('fi_incdnt_id'),
    'inner',
)

print('Подготовка данных завершена.')
```

## 4. Формирование отчёта

```python
print('Формирование отчёта...')

report = (
    enriched
    .select(
        # --- Поля инцидента
        'incdnt_id', 'incdnt_sid',
        'incdnt_status_name', 'incdnt_autoreg_flag',
        'incdnt_detection_person_name', 'incdnt_source_name',
        'src_type_lvl_1_name', 'src_type_lvl_2_name',
        'incdnt_type_lvl_1_name', 'incdnt_type_lvl_2_name',
        'incdnt_detection_dt', 'incdnt_start_dt', 'incdnt_entry_dt',
        'incdnt_first_validated_dttm', 'incdnt_last_validate_dttm',
        'risk_profile_id', 'risk_profile_name',
        'incdnt_client_type_name', 'incdnt_mistake_cnt',
        'incdnt_appl_num', 'incdnt_agr_num', 'incdnt_agr_sid',
        F.regexp_replace(F.col('incdnt_summary_descr_txt'), _UNICODE_GARBAGE, '').alias('incdnt_summary_descr_txt'),
        F.regexp_replace(F.col('incdnt_full_descr_txt'), _UNICODE_GARBAGE, '').alias('incdnt_full_descr_txt'),
        'org_struct_id',
        'org_struct_lvl_2_name', 'org_struct_lvl_3_name', 'org_struct_lvl_4_name',
        'org_struct_lvl_5_name', 'org_struct_lvl_6_name', 'org_struct_lvl_7_name',
        'org_struct_lvl_8_name', 'org_struct_lvl_9_name', 'org_struct_lvl_10_name',
        'funct_block_id',
        'funct_block_lvl_2_name', 'funct_block_lvl_3_name', 'funct_block_lvl_4_name',
        'process_lvl_1_name', 'process_lvl_2_name', 'process_lvl_3_name', 'process_lvl_4_name',
        'clntpth_lvl_4_name',
        'busn_area_id', 'busn_area_lvl_1_name', 'busn_area_lvl_2_name',
        'incdnt_security_risk_flag', 'incdnt_infrmtn_sys_risk_flag',
        'incdnt_behavior_risk_flag', 'incdnt_model_risk_flag',
        # --- Финансовое последствие (детализация)
        'fin_impact_id', 'fin_impact_sid',
        'fin_impact_kind_name',
        'fin_impact_monitoring_flag',
        'fin_impact_crncy_code', 'fin_impact_local_crncy_code',
        'fin_impact_detection_dt', 'fin_impact_creation_dttm', 'fin_impact_reg_dt',
        'fin_impact_account_num', 'fin_impact_docum_num',
        'fin_impact_ccy_amt', 'fin_impact_local_ccy_amt', 'fin_impact_rub_amt',
    )
    .dropDuplicates(['incdnt_sid', 'fin_impact_sid'])
    .orderBy('incdnt_entry_dt', 'incdnt_sid', 'fin_impact_sid')
)

print('Отчёт сформирован.')
```
